##### Linear Regression is the oldest, simple and widely used supervised machine learning algorithm for predictive analysis. 

### What is Linear Regression?

It‚Äôs a method to predict a target variable by fitting the best linear relationship between the dependent and independent variable.

We find the best fitted shape ( for two dimensions it will be a line , for three dimensions it will be plane and for other higher dimensions, a hyperplane ) for the given data. Lets first give you an intuition about Linear Regression with an example.

Let say there is a colony and there are lots of houses. And they are diverse, like few of them have a single floor, while few have 2 floors, some may have a garden , some may have parking and many other things. Let‚Äôs call these mentioned things like features a house can possess.So the price of the house will depend upon these features.

If a house has all the features then it will be costly while having only a few of them or say a single feature then it will cost less ( Still a lot though ! ). So these features are independent variables like  X<sub>1</sub>,ùëã<sub>2</sub>,ùëã<sub>3</sub>....  and the price of the house is the dependent variable as Y.

Now let‚Äôs say there are 1000 houses ( it seems to be a very big big colony :P ) and you know prices of 850 houses ( let say from House No. 1 to House No. 850 ). Now as a curious person you want to know the price of House no. 851. So what you will do? In this sort of situation generally, people make a guess that okay this house has 2 floors and garden as well so this may cost around XYZ. So you will also make a guess, but your guess will be dependent/affected by the other 850 houses whose price you know. Like if you have seen a similar house (let say house no 150 ) so you will predict the price similar to that one of House no 150 for House No. 850.

Since it is a tedious job as well as your prediction might not be correct. So here the concept of Linear Regression was introduced. Basically what we do in this is we find a linear relationship between the dependent and independent variables and then after getting that relation, we use it to find/predict the outcome of some given data.

In the above colony case, first, you will find a relation between price and different features of a house. And then using that relationship you will find the price of the remaining 150 houses.

Now how to find that relationship and what are different methods for that is explained below in brief.

The best fit is done by making sure that the sum of all the distances between the shape and the actual observations at each point is as small as possible. The fit of the shape is ‚Äúbest‚Äù in the sense that no other position would produce less error given the choice of shape.
